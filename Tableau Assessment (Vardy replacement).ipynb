```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
df_champ = pd.read_excel('D:\Tableau Assessment\Championship 22-23 Player Stats.xlsx', 'Minutes filtered')
```

```python
df_italy = pd.read_excel('D:\Tableau Assessment\Serie A 22-23 Player Stats.xlsx', 'Minutes filtered')
```

```python
df_france = pd.read_excel('D:\Tableau Assessment\Ligue 1 - 22-23 Player Stats.xlsx', 'Minutes filtered')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df = pd.concat([df_champ, df_italy, df_france]).reset_index()
df
```

```python
normalized_cols = ['Goals', 'xG', 'Assists', 'xA','Successful defensive actions per 90', 'Aerial duels per 90', 'Aerial duels won, %', 'Goals per 90', 'Non-penalty goals', 'Non-penalty goals per 90', 'xG per 90', 'Head goals', 'Head goals per 90', 'Shots', 'Shots per 90', 'Shots on target, %', 'Goal conversion, %', 'Assists per 90', 'Crosses per 90', 'Accurate crosses, %', 'Dribbles per 90', 'Successful dribbles, %', 'Offensive duels per 90', 'Offensive duels won, %', 'Touches in box per 90', 'Received passes per 90', 'Received long passes per 90', 'Fouls suffered per 90', 'Passes per 90', 'Accurate passes, %', 'Forward passes per 90', 'Accurate forward passes, %', 'xA per 90', 'Key passes per 90', 'Passes to final third per 90', 'Penalties taken', 'Penalty conversion, %']
df1 = df[normalized_cols].apply(zscore)
df1
```
